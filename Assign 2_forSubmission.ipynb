```python
import random
import math
from random import choice
from random import sample
```

```python
lone = "owning of my loneliness and the end of the loaning of my ownliness"
lonely = lone.split()
print(lonely)
```

```python
# https://www.geeksforgeeks.org/removing-stop-words-nltk-python/

import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
stop_words = set(stopwords.words('english'))
```

```python
filtered_alone = [w for w in lonely if not w in stop_words] 
print(filtered_alone) 
```

```python
filtered_stop = [w for w in lonely if w in stop_words]
print(filtered_stop)
```

```python
from IPython.display import display, HTML
def show_html(src):
    return display(HTML(src), metadata=dict(isolated=True))
```

```python
html_tmpl = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="utf-8">
    <title>{title}</title>
    <style>
    html {{ min-height: 32em; overflow: hidden; }}
    </style>
</head>
<body>
{content}
</body>
</html>"""
```

```python
content = ""
for i in range(20):
    # pick words for this story
    choosing = random.sample(filtered_alone,3)
    joined = ' '.join(random.sample(filtered_alone,3))
    alone = random.choice(filtered_alone)
    halt = random.sample(filtered_stop,2)
    halted = ' '.join(halt)
    stop = random.choice(filtered_stop)
    # concatenate together a series of f-strings...
    #story = f"<h2>{joined.capitalize()}</h2>\n"
    story = f"<h2>{joined}</h2>\n"
    story += f"<p>{halted}</p>"
    story += f"<p>{stop}</p>"
    story += f"<p>{alone}</p>"
    story += "\n"
    # and then add to the content
    content += story

divs = []
for i in range(100):
    x = random.random() * 100
    y = random.random() * 100
    this_div = f"<div style='position: absolute; left: {x}%; top: {y}%;'>{content}</div>"
    divs.append(this_div)
html_src = html_tmpl.format(title=" ", content="".join(divs))
show_html(html_src)
```

```python
open("lone.html", "w").write(html_src)
```
